#Include Dataset

```
%%capture
#@markdown <br><center><img src='https://avatars0.githubusercontent.com/u/58118658?s=400&v=4' height="140" alt="Gdrive-logo"/></center>
#@markdown <center><h3>Intiallize Pycaret </h3></center><br>
DATASET = "boston" #@param {type:"string"}
Target = "medv" #@param {type:"string"}

!pip install pycaret;
from pycaret.datasets import get_data;
dataS = get_data(DATASET); # DATASET IMPORT
print("Dataset Succesfully Imported -> Shape : ",dataS.shape)
```

#SetUp

```
#@markdown <center><h3> SEED Value  </h3></center><br>
#@markdown <center><h> Use any </h></center><br>
Set_Config = True #@param {type:"boolean"}
val=999 #@param {type:"number"}
#@markdown <hr><br>
Session_Id = False #@param {type:"boolean"}
Sval=1 #@param {type:"number"}
```

```
#@markdown <center><h1>Build Model</h1></center><br>
#@markdown <center><h>Check For Datatype and Press Enter </h></center><br>
Train_Size = 0.7 #@param {type: "slider", min: 0.1, max: 0.9, step: 0.1}

Remove_Duplicates = True #@param {type:"boolean"}
if Remove_Duplicates:
  dataS.drop_duplicates()

#@markdown <hr><br>
#@markdown <center><h>Normalize</h></center><br>
Normalize = False #@param {type:"boolean"}
Normalize_method = 'zscore'  #@param ["zscore", "minmax","maxabs", "robust"]

#@markdown <hr><br>
#@markdown <center><h>Feature Selection</h></center><br>
Feature_Selection = False #@param {type:"boolean"}
Feature_Selection_Threshold = 0.8 #@param {type: "slider", min: 0.1, max: 1.0,step: 0.1}
 

#@markdown <hr><br>
#@markdown <center><h>Outliers Removal</h></center><br>
Remove_Outliers = False #@param {type:"boolean"}
Outliers_Threshold = 0.05 #@param {type: "slider", min: 0.00, max: 0.90,step: 0.01}

#@markdown <hr><br>
#@markdown <center><h>Transformation</h></center><br>
Transformation = False #@param {type:"boolean"}
Transformation_Method ='yeo-johnson' #@param['yeo-johnson','quantile']

#@markdown <hr><br>
#@markdown <center><h>Principal Component Analysis</h></center><br>
PCA = False #@param {type:"boolean"}
Pca_Method = 'linear' #@param['linear','kernel','incremental']

if Session_Id:
  from pycaret.regression import *
  reg = setup(data = dataS, target=Target ,train_size=Train_Size ,
              normalize=Normalize ,normalize_method=Normalize_method,
              feature_selection=Feature_Selection,feature_selection_threshold=Feature_Selection_Threshold,
              remove_outliers=Remove_Outliers ,outliers_threshold=Outliers_Threshold,
              transformation=Transformation ,transformation_method=Transformation_Method,
              pca=PCA ,pca_method=Pca_Method,
              session_id = Sval
              )
else:
  from pycaret.regression import *
  reg = setup(data = dataS, target=Target ,train_size=Train_Size ,
              normalize=Normalize ,normalize_method=Normalize_method,
              feature_selection=Feature_Selection,feature_selection_threshold=Feature_Selection_Threshold,
              remove_outliers=Remove_Outliers ,outliers_threshold=Outliers_Threshold,
              transformation=Transformation ,transformation_method=Transformation_Method,
              pca=PCA ,pca_method=Pca_Method
              )
if Set_Config:
  set_config('seed',val)


```

#Best Model

```
#@markdown <center><h1>Best Model</h1></center><br>

Fold = 10 #@param {type: "slider", min: 2, max: 45}
Round_Up = 4 #@param {type: "slider", min: 1, max: 10}
Sort = 'R2' #@param['MAE','MSE','RMSE','R2','RMSLE','MAPE']
Result = compare_models(fold = Fold,  round = Round_Up ,sort=Sort);
Result = pull()
```
